Question 1: What is a random variable in probability theory?

Answer:
A random variable is a numerical quantity whose value depends on the outcome of a random experiment. Instead of describing outcomes in words (like ‚Äúheads‚Äù or ‚Äútails‚Äù), random variables assign numbers to those outcomes.

Example: If you flip a coin, you can define a random variable X such that X = 1 for heads and X = 0 for tails.

Question 2: What are the types of random variables?

Answer:
There are two main types of random variables:

Discrete Random Variable ‚Äì Takes only a finite or countably infinite number of distinct values. Example: Number of students present in a class.

Continuous Random Variable ‚Äì Takes values from an interval of real numbers, and the possible outcomes cannot be listed individually. Example: Height of a person (which can be 160.5 cm, 160.55 cm, etc.).

Question 3: Explain the difference between discrete and continuous distributions.

Answer:

Discrete Distribution:

Associated with discrete random variables.

Probabilities are assigned to specific values.

Represented by a probability mass function (PMF).

Example: Binomial distribution, Poisson distribution.

Continuous Distribution:

Associated with continuous random variables.

Probabilities are assigned over intervals, not individual points.

Represented by a probability density function (PDF).

Example: Normal distribution, Exponential distribution.

Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Answer:
The Central Limit Theorem (CLT) states that when independent random samples are taken from any population with a finite mean and variance, the sampling distribution of the sample mean will approach a normal distribution as the sample size becomes large, regardless of the original population‚Äôs distribution.

Question 7: What is the significance of confidence intervals in statistical analysis?

Answer:
A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (like mean or proportion) with a specified level of confidence (e.g., 95%).

Significance:

Provides an estimate of uncertainty around sample statistics.

Helps in decision-making by showing the reliability of the results.

Narrower intervals ‚Üí more precise estimates; wider intervals ‚Üí less precise.

Commonly used in hypothesis testing and reporting research findings.

 Example: If the 95% CI for average height = [165, 175] cm, it means we are 95% confident that the true population mean lies within this range.

```python
#Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.


#Answer:

import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from a normal distribution (mean=50, std=5)
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

print("Sample Mean:", mean)
print("Sample Standard Deviation:", std_dev)

# Draw histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black', density=True)
plt.title("Histogram of Normally Distributed Data (mean=50, std=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import scipy.stats as st

# Daily sales data (sample)
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to NumPy array
data = np.array(daily_sales)

# Sample size, mean, and standard deviation
n = len(data)
mean = np.mean(data)
std_dev = np.std(data, ddof=1)  # use ddof=1 for sample std

# 95% confidence interval using CLT (Z-distribution)
confidence_level = 0.95
alpha = 1 - confidence_level
z_critical = st.norm.ppf(1 - alpha/2)  # 1.96 for 95%

# Margin of error
margin_of_error = z_critical * (std_dev / np.sqrt(n))

# Confidence interval
ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error

print("Sample Mean:", mean)
print("Sample Standard Deviation:", std_dev)
print(f"95% Confidence Interval for Mean: ({ci_lower:.2f}, {ci_upper:.2f})")
```
